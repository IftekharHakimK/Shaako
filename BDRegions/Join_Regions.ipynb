```python
import pandas as pd

df_divisions = pd.read_csv('divisions.csv')


print(df_divisions) 
```

```python
df_divisions.drop(['div_name_en', 'site'], axis=1, inplace=True)
```

```python
df_divisions
```

```python
df_districts = pd.read_csv('districts.csv')


print(df_districts) 
```

```python
df_districts.drop(['dist_name_en', 'lat', 'lon', 'site'], axis=1, inplace=True)
```

```python
df_districts
```

```python
df_divisions.join(df_districts.set_index('div_id'), on='div_id')
```

```python
df_upazillas = pd.read_csv('upazilas.csv')


print(df_upazillas) 
```

```python
df_upazillas.drop(['upazilla_name_en', 'site'], axis=1, inplace=True)
```

```python
df_upazillas
```

```python
df_unions = pd.read_csv('unions.csv')


print(df_unions) 
```

```python
df_unions.drop(['union_name_en', 'site'], axis=1, inplace=True)
```

```python
df_unions
```

```python
df_combined = df_divisions.join(df_districts.set_index('div_id'), on='div_id')
```

```python
df_combined
```

```python
df_combined = df_combined.join(df_upazillas.set_index('dist_id'), on='dist_id')
```

```python
df_combined
```

```python
df_combined = df_combined.join(df_unions.set_index('upazilla_id'), on='upazilla_id')
```

```python
df_combined
```

```python
df_combined.drop(['div_id', 'dist_id', 'upazilla_id', 'union_id'], axis=1, inplace=True)
```

```python
df_combined
```

```python
df_combined.to_csv('regions.csv', index=False)
```
